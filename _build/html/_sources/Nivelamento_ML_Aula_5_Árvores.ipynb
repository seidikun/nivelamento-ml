Árvore, em programação, é uma estrutura que contêm nós e conexões entre eles. Pense numa árvore mesmo

Vamos à definição formal desse modelo: árvores, de um modo geral em teoria da computação, são estruturas de dados formadas por um conjunto de elementos que armazenam informações chamadas **nós** e a **junção entre dois nós é feito por meio de uma aresta**. Além disso, um nó que não possui filho, ou seja, não se conecta a nenhum nó num estágio inferior é chamado de **folha**. Abaixo temos o exemplo de uma árvore simples (de ponta-cabeça, lembra?).

Resumindo:

- Árvore é uma estrutura de dados composta de nós e arestas;
- **Nó** é a estrutura que armazena informação;
- **Aresta** é a estrutura que interconecta 2 nós;
- O primeiro nó é conhecido como **raiz**;
- Um **nó folha** é aquele que não possui nenhum filho;
- A **profundidade** de uma árvore é quantos nós preciso passar para ir da raiz até a folha mais distante (no caso acima, a profundidade é **5**).

Além disso, para compreendermos como funcionam Árvores de decisão mais um conceito ainda precisa ser apresentado: **Árvores Binárias**. Por definção:
> Uma árvore binária é um tipo especial de árvore que obedece às seguintes condições:
> - Não tem elemento algum (árvore vazia).
> - Ou um nó possui 0 filhos ou 2 filhos distintos.
