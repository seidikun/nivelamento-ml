## Uma definição mais concreta

A lógica de programação é uma forma de pensar e descrever comandos ao computador, a partir das ações básicas que ele consegue executar. O computador é pensado há muitos anos, mas o modelo que se estabeleceu é o modelo de Alan Turing (1912 - 1954).


alan_turing_header.jpg


No modelo computacional de Turing, ações básicas como aritmética (soma, subtração, multiplicação, divisão), comandos de acesso a endereços de memória, condicionais ('se', 'senão') e repetição ('enquanto') são usadas para realizar todo tipo de operação mais complexa. Esse mesmo modelo é o que estabelece as operações físicas do computador moderno e que baseia qualquer linguagem de programação.




````python
# @title Vídeo: Máquina de Turing
from ipywidgets import widgets
from IPython.display import YouTubeVideo
from IPython.display import IFrame
from IPython.display import display

def display_videos(video_ids, W=400, H=300, fs=1):
  tab_contents = []
  for i, video_id in enumerate(video_ids):
    out = widgets.Output()
    with out:
      video = YouTubeVideo(id=video_ids[i][1], width=W,
                            height=H, fs=fs, rel=0)
      print(f'Video available at https://youtube.com/watch?v={video.id}')
      display(video)
    tab_contents.append(out)
  return tab_contents


video_ids = [('Youtube', 'dNRDvLACg5Q')]
tab_contents = display_videos(video_ids, W=854, H=480)
tabs = widgets.Tab()
tabs.children = tab_contents
for i in range(len(tab_contents)):
  tabs.set_title(i, video_ids[i][0])
display(tabs)
````

Da mesma forma que os comandos gerados para o carrinho no exercício acima, uma sequência lógica de comandos para o computador faz com que ele realize todo tipo de ação, de mostrar imagens na tela, acessar a internet, tocar música a realizar análises e fazer descobertas.

Alan Turing, mesmo no início da computação, já pensava em como um sistema de ações simples poderia eventualmente gerar pensamentos e criar ideias. Por conta disso, ele é conhecido não apenas como o pai da computação, mas também da inteligência artificial.

## Algoritmo

Nesse ponto, você consegue dizer o que é um algoritmo? Esse conceito é a base da computação, afinal fazer computação é desenvolver algoritmos.

Podemos pensar em algoritmos como receitas de bolo: uma sequência de ações (preparos) sobre variáveis (ingredientes) realiza um objetivo (fazer o bolo!). Essas ações, no computador, seguem a lógica da máquina de Turing, ou seja, são ações muito simples, e inicialmente a programação era realizada determinando cada passo simples a ser realizado pela máquina. Um exemplo abaixo:


```
Algoritmo 'faça um bolo de laranja'

Variáveis: 2 laranjas, 2 ovos, 1 xícara de óleo, 1 xícara e meia de açúcar, 2 xícaras de farinha de trigo, 2 colheres de fermento em pó

Ações
- Descasque as laranjas e retire as sementes
- Adicione as laranjas, ovos, óleo e açúcar no recipiente do liquidificador
- Bata a mistura no liquidificador por 2 minutos
- Numa tigela, misture a farinha e fermento em pó
- Transfira o conteúdo do liquidificador para a tijela e misture levemente para incorporar os ingredientes secos
- Transfira o conteúdo todo para uma assadeira untada
- Leve a assadeira para o forno pré-aquecido a 180º por 30 minutos
```

**[Exercício] Escreva em suas palavras o que é um algoritmo**

Com o tempo, os programadores observaram que era possível abstrair sequências específicas de ações em ações maiores. Um exemplo: em vez de ensinar o passo-a-passo do 'algoritmo' de como fazer um bolo de laranja ao computador, escreve-se esse algoritmo uma vez e podemos chamá-lo de 'faça um bolo de laranja'.

Obs.: na sequência de ações do algoritmo 'faça um bolo de laranja', é possível argumentar que os passos descritos também são algoritmos em si. Por exemplo 'descasque as laranjas' é um processo físico que envolve etapas mais simples como pegar um descascador, segurar a laranja, posicionar o descascador, etc.

Podemos pensar em algoritmos 'faça um bolo de laranja', 'faça um cafezinho' e muito mais!
