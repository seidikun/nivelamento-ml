### Subplots [Troca de apresentador/a]

Ey34r4dWgAwZ1nb.png


A utilização de subplots é interessante quando queremos criar mais de um gráfico na mesma figura. Para apresentar esse tópico, utilizaremos um histograma (plot estatístico) da quantidade de vagas e um gráfico de pizza (plot categórico) da "idade" (apartamento novo ou não).

Existem muitas formas de criar os subplots. Vamos apresentar um exemplo com a função `subplot`, que pode ser consultada [na documentação](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html). Se preferir, outra função útil é `subplots`, que também pode ser vista [na documentação](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html).

````python
fig = plt.figure(figsize = (10, 4))

plt.subplot(1, 2, 1) # o primeiro 1: qntd de linhas, 2 = qntd de colunas, ultimo 1 = primeiro gráfico

plt.hist(df_aluguel["vagas"]) # .hist = histograma
plt.title("Histograma: Vagas")

plt.subplot(1, 2, 2)
#  conta quantos sim/não tem de ap novo    | primeiro valor é sim, segundo é não
plt.pie(df.groupby("novo")["valor"].count(), labels = ["sim", "não"])
plt.title("Pizza: Novo")

fig.suptitle("Exemplo de subplot")

plt.show()
````

### Gráficos simples 

Vamos começar vendo como fazer alguns plots simples usando o seaborn

#### Gráfico de barras

Gráficos de barras podem ser usados para visualizarmos tanto dados categóricos quanto numéricos. 

Para fazer um gráfico de barras com o seaborn podemos usar a função `.countplot()`, que é um gráfico de contagem simples. Para isso, definimos no parâmetro da função qual coluna será o eixo x (o que queremos "contar") e também de onde vamos tirar esses dados, especificando o parâmetro "data". 

Portanto, o modelo para criar um countplot simples é:

```sns.countplot(x = 'coluna', data = nome_do_dataframe)```

Aqui vamos ver um gráfico simples da contagem de quartos:

````python
sns.countplot(x = 'quartos', data = df) # data = nome do dataframe 
plt.show()
````

No gráfico acima podemos ver que muitas casas possuem cerca de 2 e 3 quartos. E que muitas poucas possuem apenas 1 quarto ou mais do que 4.

**[Exercício] Agora vamos ver a contagem de elevador:**


````python
# Ao definir o parêmetro "palette" podemos mudar a cor da peleta de cores do gráfico
sns.countplot(x = 'elevador', data = df, palette='rocket') # palette = cor
plt.show()
````

No gráfico acima, por sua vez, podemos ver que a maioria das casas possuem cerca de 2 banheiros, sendo a maior parte concentrada entre 1 e 2. Muitas poucas casas possuem mais do que 3 banheiros.

### Gráficos com duas variáveis [Troca de apresentador/a]

Até agora vimos como adicionar uma das colunas do nosso dataframe ao gráfico, mas podemos adicionar duas e até mais ao criarmos um gráfico. Vamos ver como fazer isso com gráficos simples:

#### Gráfico de ponto / distribuição

Gráficos de ponto (scatterplot) podem ajudar a entender a distribuição dos nossos dados, e até mesmo enxergar se há correlação entre eles.

Podemos usar o `scatterplot` do seaborn para plotarmos esse tipo de gráfico, sendo os parâmetros principais:

`sns.scatterplot(x ='nome da coluna', y = 'nome da coluna', data = nome_do_df`)

**[Exercício] Vamos começar vendo um gráfico de pontos que relaciona o valor e as vagas:**


````python
sns.scatterplot(x= 'valor', y= 'vagas', data=df)
plt.show()
````

Com o gráfico pequeno assim é meio díficil de visualizar bem os pontos, então que tal aumentarmos um pouco a largura dele? Para isso utilizamos o `figsize` do matplotlib antes do código do gráfico em si, ele funciona assim:

`plt.figure(figsize(largura, altura))`


Então vamos ver como o gráfico acima fica modificando o tamanho:



````python
plt.figure(figsize=(15,5))

sns.scatterplot(x= 'valor' , y= 'vagas', data=df)
plt.show()
````

Dessa forma podemos ver que a maior concentração de dados está entre 1 e cerca de 300 m².


Também podemos usar o `pointplot` do seaborn para criarmos gráficos de pontos com linhas para "conectá-los". Ele funciona de forma semelhante ao que já vimos!

`sns.pointplot(x = 'nome da coluna', y = 'nome da coluna', data = nome_df)`

Vamos ver como funciona:

````python
# aqui alterei o tamanho do gráfico para ele ficar mais legível
plt.figure(figsize=(17,5))

sns.pointplot(x='vagas', y = 'tamanho', data = df)
plt.show()
````

Caso você queira que os pontos não estejam ligados, basta adicionar o parâmetro `join`igual a False, da seguinte forma:


````python
# aqui alterei o tamanho do gráfico para ele ficar mais legível
plt.figure(figsize=(17,5))

sns.pointplot(x='vagas', y = 'tamanho', data = df, join = False)
plt.show()
````

### Gráfico de caixa

Neste gráfico 
* A caixa colorida representa o porcentil
* A linha no meio representa a mediana 
* As linhas do lado de fora dão uma ideia da distribuição, onde é o valor máximo e onde é o valor minímo
* E os pontos flutuantes representam valores extremos (ouliers). 

Para uma imagem ilustrativa veja [este link](https://i0.wp.com/www.abgconsultoria.com.br/blog/wp-content/uploads/boxplot-1-1024x664.png?resize=1024%2C664).
Para criar o gráfico podemos usar o `boxplot`do seaborn:

````python
plt.figure(figsize = (15, 8))
sns.boxplot(x='quartos', y = 'tamanho', data=df, palette = "gist_rainbow")
plt.show()
````
