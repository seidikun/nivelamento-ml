Criando uma classe carro

A classe carro possui:

**Atributos**


*   cor
*   marca
*   tipo

**Funções**



*   iniciar_carro()
*   descrever()
*   acelerar()
*   parar()





```python
class Carro:
  def __init__(self):
    self.cor        = ''
    self.marca      = ''
    self.tipo       = ''
    self.velocidade = 0

  def iniciar_carro(self, cor, marca, tipo):
    ''' inicia um carro '''
    self.cor        = cor
    self.marca      = marca
    self.tipo       = tipo
    self.descrever()

  def descrever(self):
    ''' descreve os atributos do carro '''
    print('O carro é', self.cor, 'da marca', self.marca, 'e tipo', self.tipo)

  def acelerar(self, valor):
    ''' aumenta a velocidade do carro '''
    self.velocidade = valor
    print('O carro está a', self.velocidade, 'km/h')

  def parar(self):
    ''' diminui a velocidade do carro para 0 '''
    self.acelerar(0)
```

Instanciando um carro meu_carro

```python
meu_carro = Carro()
print(meu_carro.cor)
```

```python
help(meu_carro.iniciar_carro)
```

```python
meu_carro.iniciar_carro('amarelo', 'VW', 'sedan')
```

```python
meu_carro.acelerar(30)
```

```python
print(type(meu_carro))
```

Instanciando outro carro, seu_carro

```python
seu_carro = Carro()
seu_carro.iniciar_carro('preto', 'Peugeot', 'compacto')
```

```python
seu_carro.cor = 'azul'
```

```python
seu_carro.descrever()
```

```python
seu_carro.acelerar(60)
```
