## Capitulo 2: Criando e Manipulando um dataset

### Cria√ß√£o de dataframes

Podemos criar dataframes a partir de diferentes estruturas, como:
* Lista de dicion√°rios
* Dicion√°rio de listas
* Lista de listas

Outras op√ß√µes podem ser vistas [aqui](https://www.geeksforgeeks.org/different-ways-to-create-pandas-dataframe/).

````python
# Usando lista de dicion√°rios para criar o DataFrame
import pandas as pd

pessoas = [{"nome": "Vitoria", "idade": 22, "curso": "Lingu√≠stica"},
           {"nome": "Rafael", "idade": 20, "curso": "Engenharia El√©trica"},
           {"nome": "Nelson", "idade": 20, "curso": "Matem√°tica Aplicada"},
           {"nome": "Noel", "idade": 23, "curso": "Administra√ß√£o"}]

# Atribuir o DataFrame √† vari√°vel "df_exemplo"
df_exemplo = pd.DataFrame(pessoas)

# Exibir o DataFrame ap√≥s a defini√ß√£o do √≠ndice
df_exemplo
````

### Manipulando os √≠ndices do dataframe

Tamb√©m podemos criar dataframes com √≠ndices personalizados

````python
# Utiliza a op√ß√£o "index" para definir os r√≥tulos dos √≠ndices
df_exemplo2 = pd.DataFrame(pessoas, index=["Pessoa 1", "Pessoa 2", "Pessoa 3", "Pessoa 4"])

# Exibir o DataFrame ap√≥s a defini√ß√£o do √≠ndice
df_exemplo2
````

Tamb√©m podemos definir uma coluna especifica do dataframe para ser o √≠ndice

````python
# Definir a coluna "nome" como √≠ndice das linhas
df_exemplo.set_index("nome", inplace=True)

# Exibir o DataFrame ap√≥s a defini√ß√£o do √≠ndice
df_exemplo
````

2. Encontrar o maior valor de um determinado atributo (coluna) - fun√ß√£o max()

````python
df["valor"].max()
````

````python
# Salvando o maior quantia dentro da coluna / atributo "valor"
max_valor = df["valor"].max()

# Realizando um filtro / busca em cima da v√°riavel "max_valor"
# Atualmente √© uma busca por amostras com esse valor. Mas tente substituir o "==" por ">" ou "<":
df[df["valor"] == max_valor]
````

3. Uma possibilidade de procesamento √© reduzirmos a quantidade de colunas criando um novo dataframe, e em cima desse conjunto derivado filtrarmos o que a gente deseja: 

(A redu√ß√£o de colunas ou amostras √© uma forma de otimizarmos o tempo e o uso de recursos no processamento dos dados.)

````python
# Criando um novo dataframe(df_vendas) ao adicionar as amostras do dataframe orifinal(df) que na coluna "tipo" (df["tipo"]) possuir o valor "venda"
df_vendas = df[df["tipo"] == "venda"]

# E com a fun√ß√£o "head()" podemos visualizar essa nova contru√ß√£o de maneira r√°pida
df_vendas.head()
````

Perceba que na estrutura representada a cima (dataset df_vendas) quando usamos o m√©todo "head()" percebemos um padr√£o. As 5 primeiras linhas do "df_vendas" come√ßam da linhas 4901 do dataset original. Isso n√≥s motra duas informa√ß√µes interessantes:

1¬∫ Quando pegamos linhas de um dataframe para criarmos um novo. O novo n√£o zera a numera√ß√£o antiga das linhas. 

2¬∫ A partir desse dataset, em especifico, podemos gerar uma hip√≥tese que o dataset original foi organizado ou criado artificialmente para que os "Apartamentos a venda" s√≥ aparecessem a partir da linha 4901.

4. Encontrar o menor valor de um determinado atributo (coluna) - fun√ß√£o min()

````python
# Salvando o menor quantia dentro da coluna / atributo "valor" em um dataset s√≥ formado por apartamentos que est√£o a venda
min_valor = df_vendas['valor'].min()

# Ao ver o sinal de "==" percebemos que esse algoritmo est√° tentando realizar uma busca por amostras que possuem esse "min_valor"
df_vendas[df_vendas['valor'] == min_valor]
````

5. Voltando ao dataset original: Podemos calcular a mediana dos valores de todas as colunas das amostras onde na coluna "elevador" o valor √© "sim". (O apartamento possui um elevador)

# A fun√ß√£o "median()" serve para calcular / visualizar a mediana de todos os valores das colunas numericas do dataframe 
df_vendas[df["elevador"] == "sim"].median()

````python
df_vendas[df_vendas["elevador"] == "sim"]
````

Claro que podemos adicionar mais valores para especificar ainda mais a visualiza√ß√£o dos dados presentes na saida desse filtro, como √© o caso de querermos ver somente a mediana da coluna "suites":
